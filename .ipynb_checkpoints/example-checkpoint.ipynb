```python
import os 
import matplotlib.pyplot as plt
import networkx as nx
import pickle
import itertools
resultBase = "..."
files = os.listdir(resultBase)
from pylab import rcParams
rcParams['figure.figsize'] = 20, 15
files[0]
```

```python
# draw a graph using the names coming from the EcoNameTranslator
def drawGraph(results):
    results = list(map(lambda x: x[0],results))
    test = set(itertools.chain(*results))
    if len(test) < 60 and len(test) > 20:
        G = nx.DiGraph()
        for pred,prey in results:
            G.add_edge(pred,prey)
        pos=nx.spring_layout(G,k=1)
        for label in pos:
            x,y = pos[label]
            plt.text(x,y+0.035,s=label,horizontalalignment='center',fontsize=12)
        nx.draw(G,pos,node_size=170,arrowsize=20,arrowstyle='-|>',node_color='blue')
        return True
    return False

## Draws the re-translated version of the graph, that uses original input labels
def drawGraphOrg(results):
    results = list(map(lambda x: x[1],results))
    test = set(itertools.chain(*results))
    print(len(test))
    if len(test) < 60 and len(test) > 20:
        G = nx.DiGraph()
        for pred,prey in results:
            G.add_edge(pred,prey)
        pos=nx.spring_layout(G,k=1)
        for label in pos:
            x,y = pos[label]
            plt.text(x,y+0.035,s=label,horizontalalignment='center',fontsize=12)
        nx.draw(G,pos,node_size=170,arrowsize=20,arrowstyle='-|>',node_color='blue')
        print(G.edges())
        return True
    return False
```

```python
with open(resultBase+"/BaseCSV#texasGrasslands",'rb') as f:
    arr = pickle.load(f)
    result,actual,allInteractions,finalTime,translatedTime,done,notDone = arr[0]
    if (drawGraphOrg(result)):print("OO")
```
