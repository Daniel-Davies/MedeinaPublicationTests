# Get results generated by Medeina for a given dataset

Results are formatted as as a list of tuples, where each tuple is an interaction between two species hypothesised by Medeina in the original dataset

## Result 1

We use the reef_spnames dataset from dryad (see the dataset_list for more citation details)


```python
import pickle

with open(resultFamily+"/validation#reef_spnames",'rb') as f:
    rs = pickle.load(f)

results = list(map(lambda x: x[1],rs[0][0]))
```
