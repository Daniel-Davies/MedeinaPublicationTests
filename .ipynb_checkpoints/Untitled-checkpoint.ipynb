```python
def findDuplicates(path):
    allInteractionDat = []
    files = os.listdir(path)
    cnt = 0
    counter = 0
    for item in files:
        with open(path+"/"+item,'rb') as f:
            arr = pickle.load(f)
            result,actual,allInteractions,finalTime,translatedTime,done,notDone = arr[0]
            
            allInteractions = set(sorted(sortTuples(allInteractions)))
            allInteractionDat.append((allInteractions,item))
    t = 0
    its = set()
    for k1,(item1,id1) in enumerate(allInteractionDat):
        for k2,(item2,id2) in enumerate(allInteractionDat):
            if k1 != k2:
                if len(item1) == len(item2):
                    same = True
                    ti1 = sorted(list(item1))
                    ti2 = sorted(list(item2))
                    for i in range(len(item1)):
                        if ti1[i] != ti2[i]:
                            same = False
                            break
                    if same:
                        t+=1
                        print(id1,id2)
                        its.add(sortSingleTuple((k1,k2)))
    print(t)
    print(len(its))
#         print("Pass "+str(k1)+" complete")
        
def sortTuples(setOfTuples):
    return set(map(sortSingleTuple,setOfTuples))

def sortSingleTuple(tup):
    x,y = tup
    if x < y:
        return (x,y)
    elif y < x:
        return (y,x)
    else:
        return (x,y)
```
